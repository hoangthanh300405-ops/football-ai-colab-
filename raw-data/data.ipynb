```python
# Let's get the 'homeTeamId' for the first row (index 0)
# Method 1: Direct indexing after selecting the column (returns a Series)
value_direct_index = match_df['homeTeamId'][0]
print(f"homeTeamId for row 0 (direct index): {value_direct_index}")

# Method 2: Using .loc (recommended for DataFrames) with row label and column name
value_loc = match_df.loc[0, 'homeTeamId']
print(f"homeTeamId for row 0 (.loc): {value_loc}")

# You can also select a specific row first and then the column
row_0 = match_df.loc[0]
value_from_row = row_0['homeTeamId']
print(f"homeTeamId for row 0 (from selected row): {value_from_row}")
```

```python
selected_team_id = 360 # Example teamId, you can change this

# Filter team_df to find the row with the matching teamId
team_info = team_df[team_df['teamId'] == selected_team_id]

# Check if a team was found
if not team_info.empty:
    short_display_name = team_info['shortDisplayName'].iloc[0]
    print(f"The shortDisplayName for teamId {selected_team_id} is: {short_display_name}")
else:
    print(f"No team found with teamId {selected_team_id}")
```
